### Filtragem de dados

```python
subDf = df[["maiorAtraso", "margemBrutaAcumulada", "percentualProtestos", "primeiraCompra", "prazoMedioRecebimentoVendas", "titulosEmAberto", "valorSolicitado", "valorAprovado", "percentualRisco", "capitalSocial", "totalPatrimonioLiquido", "totalAtivo", "passivoCirculante", "ativoCirculante" ]]
```

### Tratamento de objetos dateTime

```python
subDf = subDf.dropna()
subDf = subDf.drop_duplicates()
subDf['primeiraCompra'] = pd.to_datetime(subDf['primeiraCompra'], errors='coerce')
subDf['primeiraCompra'] = subDf['primeiraCompra'].dt.year
```

### Limpando colunas sem valor

```python
subDf = subDf.dropna()
```

### Separação dos dados

```python
x = subDf[["titulosEmAberto", "valorSolicitado", "percentualRisco", "capitalSocial", "totalPatrimonioLiquido", "totalAtivo", "passivoCirculante", "ativoCirculante" ]]
```

```python
y = subDf["valorAprovado"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.2, random_state=42)
```

### Treinamento do modelo


```python
modelo = LinearRegression()
```

```python
modelo.fit(x_treino, y_treino)
```

```python
predicao = modelo.predict(x_teste)
```

```python
r2_score(y_teste, predicao)
```
